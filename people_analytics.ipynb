```python
import pandas as pd
import matplotlib.pyplot as plt
#from matplotlib.animation import FuncAnimation
```

```python
# Carregando os dados da planilha
df = pd.read_excel('/content/drive/MyDrive/TechPro_Employee_Data.xlsx')
```

```python
df.head()
```

```python
# @title Department

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Department').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Ordenar os dados por data, caso haja uma coluna de tempo
df = df.sort_values('Department')
df.head()
```

```python
media_satisfacao = df['Satisfaction'].mean()
print(media_satisfacao)
```

```python
mediana_produtividade = df['Productivity'].median()
mediana_produtividade
```

```python
moda_hrs_treinamento = df['TrainingHours'].mode()[0]
print(moda_hrs_treinamento)
```

```python
std_satisfacao = df['Satisfaction'].std()
print('Desvio padro - Satisfao', std_satisfacao)
```

```python
variancia_absenteismo= df['AbsenteeismDays'].var()
print("Quantidade de Dias de Absentesmo",  variancia_absenteismo)
```

```python
amplitude_satisfacao = df['Satisfaction'].max() - df['Satisfaction'].min()
amplitude_satisfacao
```

```python
vida_trabalho = df['WorkLifeBalance'].quantile(0.75) - df['WorkLifeBalance'].quantile(0.25)
print("Equilibrio entre vida pessoal e profissional: ", vida_trabalho)
```

```python
corrmap = df.corr(method="spearman", numeric_only=True)
corrmap.describe()
```

```python
# Calculate the average satisfaction by department.
avg_satisfaction_by_department = df.groupby('Department')['Satisfaction'].mean()
print("Media de Satisfao por Departamento:\n", avg_satisfaction_by_department)

# Find the correlation between satisfaction and productivity.
correlation = df['Satisfaction'].corr(df['Productivity'])
print("\nCorrelao entre satisfao do funcionrio e a produtividade:", correlation)

# Identify the department with the highest turnover rate.
department_with_highest_turnover = df.groupby('Department')['Turnover'].mean().idxmax()
print("\nDepartamento com o maior ndice de rotatividade (turnover):", department_with_highest_turnover)

# Determine the average training hours for employees who left the company.
avg_training_hours_for_turnover = df[df['Turnover'] == 1]['TrainingHours'].mean()
print("\nMedia de horas de trainamento para funcionrios que deixaram a empresa:", avg_training_hours_for_turnover)

# Analyze the relationship between work-life balance and absenteeism.
worklifebalance_absenteeism_correlation = df['WorkLifeBalance'].corr(df['AbsenteeismDays'])
print("\nCorrelao de equilbrio entre vida profissional e pessoal e o absentesmo:", worklifebalance_absenteeism_correlation)

# Explore the impact of innovation culture on employee satisfaction.
innovation_satisfaction_correlation = df['InnovationCulture'].corr(df['Satisfaction'])
print("\nCorrelao entre a cultura de inovao e satisfao dos trabalhadores:", innovation_satisfaction_correlation)
```

```python
import matplotlib.pyplot as plt
# Grfico de barras da mdia de satisfao por departamento
plt.figure(figsize=(10, 6))
avg_satisfaction_by_department.plot(kind='bar')
plt.xlabel('Departamento')
plt.ylabel('Mdia de Satisfao')
plt.title('Mdia de Satisfao por Departamento')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
# Histograma da satisfao dos funcionrios
plt.figure(figsize=(8, 6))
plt.hist(df['Satisfaction'], bins=10)
plt.xlabel('Satisfao')
plt.ylabel('Frequncia')
plt.title('Histograma da Satisfao dos Funcionrios')
plt.show()
```

```python
# Grfico de disperso entre satisfao e produtividade
plt.figure(figsize=(8, 6))
plt.scatter(df['Satisfaction'], df['Productivity'])
plt.xlabel('Satisfao')
plt.ylabel('Produtividade')
plt.title('Relao entre Satisfao e Produtividade')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Boxplot da satisfao por departamento
plt.figure(figsize=(10, 6))
df.boxplot(column='Satisfaction', by='Department', rot=45)
plt.xlabel('Departamento')
plt.ylabel('Satisfao')
plt.title('Boxplot da Satisfao por Departamento')
plt.show()
```

```python
#gerar grfico mapa de calor

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 10))
sns.heatmap(corrmap, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor da Correlao entre as Variveis')
plt.show()
```
