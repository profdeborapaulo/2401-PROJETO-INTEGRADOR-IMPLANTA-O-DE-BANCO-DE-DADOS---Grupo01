```python
import pandas as pd
import matplotlib.pyplot as plt
#from matplotlib.animation import FuncAnimation
```

```python
# Carregando os dados da planilha
df = pd.read_excel('/content/drive/MyDrive/TechPro_Employee_Data.xlsx')
```

```python
df.head()
```

```python
# @title Department

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Department').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Ordenar os dados por data, caso haja uma coluna de tempo
df = df.sort_values('Department')
df.head()
```

```python
media_satisfacao = df['Satisfaction'].mean()
print(media_satisfacao)
```

```python
mediana_produtividade = df['Productivity'].median()
mediana_produtividade
```

```python
moda_hrs_treinamento = df['TrainingHours'].mode()[0]
print(moda_hrs_treinamento)
```

```python
std_satisfacao = df['Satisfaction'].std()
print('Desvio padrão - Satisfação', std_satisfacao)
```

```python
variancia_absenteismo= df['AbsenteeismDays'].var()
print("Quantidade de Dias de Absenteísmo",  variancia_absenteismo)
```

```python
amplitude_satisfacao = df['Satisfaction'].max() - df['Satisfaction'].min()
amplitude_satisfacao
```

```python
vida_trabalho = df['WorkLifeBalance'].quantile(0.75) - df['WorkLifeBalance'].quantile(0.25)
print("Equilibrio entre vida pessoal e profissional: ", vida_trabalho)
```

```python
corrmap = df.corr(method="spearman", numeric_only=True)
corrmap.describe()
```

```python
# Calculate the average satisfaction by department.
avg_satisfaction_by_department = df.groupby('Department')['Satisfaction'].mean()
print("Media de Satisfação por Departamento:\n", avg_satisfaction_by_department)

# Find the correlation between satisfaction and productivity.
correlation = df['Satisfaction'].corr(df['Productivity'])
print("\nCorrelação entre satisfação do funcionário e a produtividade:", correlation)

# Identify the department with the highest turnover rate.
department_with_highest_turnover = df.groupby('Department')['Turnover'].mean().idxmax()
print("\nDepartamento com o maior índice de rotatividade (turnover):", department_with_highest_turnover)

# Determine the average training hours for employees who left the company.
avg_training_hours_for_turnover = df[df['Turnover'] == 1]['TrainingHours'].mean()
print("\nMedia de horas de trainamento para funcionários que deixaram a empresa:", avg_training_hours_for_turnover)

# Analyze the relationship between work-life balance and absenteeism.
worklifebalance_absenteeism_correlation = df['WorkLifeBalance'].corr(df['AbsenteeismDays'])
print("\nCorrelação de equilíbrio entre vida profissional e pessoal e o absenteísmo:", worklifebalance_absenteeism_correlation)

# Explore the impact of innovation culture on employee satisfaction.
innovation_satisfaction_correlation = df['InnovationCulture'].corr(df['Satisfaction'])
print("\nCorrelação entre a cultura de inovação e satisfação dos trabalhadores:", innovation_satisfaction_correlation)
```

```python
import matplotlib.pyplot as plt
# Gráfico de barras da média de satisfação por departamento
plt.figure(figsize=(10, 6))
avg_satisfaction_by_department.plot(kind='bar')
plt.xlabel('Departamento')
plt.ylabel('Média de Satisfação')
plt.title('Média de Satisfação por Departamento')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
# Histograma da satisfação dos funcionários
plt.figure(figsize=(8, 6))
plt.hist(df['Satisfaction'], bins=10)
plt.xlabel('Satisfação')
plt.ylabel('Frequência')
plt.title('Histograma da Satisfação dos Funcionários')
plt.show()
```

```python
# Gráfico de dispersão entre satisfação e produtividade
plt.figure(figsize=(8, 6))
plt.scatter(df['Satisfaction'], df['Productivity'])
plt.xlabel('Satisfação')
plt.ylabel('Produtividade')
plt.title('Relação entre Satisfação e Produtividade')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Boxplot da satisfação por departamento
plt.figure(figsize=(10, 6))
df.boxplot(column='Satisfaction', by='Department', rot=45)
plt.xlabel('Departamento')
plt.ylabel('Satisfação')
plt.title('Boxplot da Satisfação por Departamento')
plt.show()
```

```python
#gerar gráfico mapa de calor

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 10))
sns.heatmap(corrmap, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor da Correlação entre as Variáveis')
plt.show()
```

### Vamos analisar cada um dos resultados acima:

### 1. **Correlação entre Satisfação e Produtividade: -0,0129**

Uma correlação de -0,0129 é praticamente zero, o que indica que não há uma relação linear significativa entre satisfação dos funcionários e produtividade. Em outras palavras, não podemos concluir que funcionários mais satisfeitos são mais produtivos ou vice-versa. Este resultado sugere que outros fatores podem estar influenciando a produtividade, e que a satisfação sozinha não é um preditor confiável para isso.

### 2. **Departamento com Maior Índice de Rotatividade: Financeiro**

O departamento financeiro apresenta o maior índice de rotatividade na organização. Esse dado pode indicar vários problemas potenciais, como insatisfação com a carga de trabalho, falta de oportunidades de crescimento, problemas de gestão, ou até mesmo fatores externos, como a demanda por profissionais de finanças no mercado. Uma análise mais aprofundada deve ser feita para identificar as causas específicas e desenvolver estratégias de retenção para esse departamento.

### 3. **Média de Horas de Treinamento para Funcionários que Saíram da Empresa: 10,38**

Este valor médio indica que os funcionários que deixaram a empresa passaram, em média, cerca de 10 horas em treinamentos antes de saírem. Isso pode sugerir que os treinamentos oferecidos podem não estar sendo efetivos em engajar e reter esses funcionários, ou que o treinamento não está alinhado com as necessidades ou expectativas deles. Alternativamente, também pode significar que o treinamento está sendo dado em um estágio tardio da jornada do funcionário, quando a decisão de sair já foi tomada.

### 4. **Correlação entre Equilíbrio entre Vida Pessoal e Profissional e Absenteísmo: -0,0086**

A correlação de -0,0086 entre equilíbrio entre vida pessoal e profissional e absenteísmo é muito próxima de zero, indicando que não há uma relação linear significativa entre essas duas variáveis. Esse resultado sugere que o equilíbrio entre a vida pessoal e profissional, conforme medido na organização, pode não ter um impacto direto no absenteísmo dos funcionários. Outros fatores, como a cultura da empresa, condições de saúde, ou problemas pessoais, podem ter um impacto maior no absenteísmo.

### 5. **Correlação entre Cultura de Inovação e Satisfação dos Funcionários: -0,0137**

A correlação de -0,0137 entre cultura de inovação e satisfação dos funcionários também é praticamente zero, indicando que a percepção de inovação na cultura da empresa não está diretamente associada à satisfação dos funcionários. Isso pode ser devido a vários motivos, como a percepção de que a inovação não está trazendo benefícios diretos para o dia a dia dos funcionários, ou que não há suporte suficiente para a inovação de fato acontecer.

- É importante lembrar que correlações fracas ou próximas de zero não significam que não há nenhuma relação entre as variáveis; elas apenas indicam que uma relação linear simples pode não ser suficiente para capturar o relacionamento entre essas variáveis.
